```python
import cv2
```

```python
img = cv2.imread('demo/demo_text_ocr.jpg')
```

```python
plt.imshow(img)
```

```python
from mmocr.apis import MMOCRInferencer
infer = MMOCRInferencer(rec='ABINet_Vision')
result = infer(img, save_vis=True, return_vis=True)
print(result['predictions'])
```

```python
from mmocr.apis import MMOCRInferencer
infer = MMOCRInferencer(det='dbnetpp')
result = infer(img, return_vis=True)
```

```python
print(result)
```

```python
import numpy as np

arr = np.array(result['predictions'][0]['det_polygons'])
print(arr.shape)
```
